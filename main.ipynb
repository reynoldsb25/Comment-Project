```python
import csv

#Opens teacher_comments csv file, saves header as header, then skips to first row of data. Then moving through the csv file line by line it appends each line to a list called comments.
comments = []
with open('teacher_comments.csv') as com:
    header = com.readline()
    data = csv.reader(com, delimiter=",")
    for line in data:
        comments.append(line)

#This is the overview of what the students did during the semester. This string is appended at the beginning of each student's file to provide the reader a summary of the course.
course = ("This semester in Computer Science 1 we have covered many topics. After learning each topic the students then used that knowledge to solve multiple questions in a problem set. Then after mastering the new skills the students would take a written test, challenging them to code in a new way. Everyday students come to class ready to learn new concepts which can be used to solve real world problems.")


def work_ethic(first, pro1, pro2, participation, motivation, attendance):
    '''This function takes in a person’s name, their personal pronouns, and different characteristics on scales from 1-10, and creates a personal paragraph commenting on how the student's behavior, motivation, and attendance has been this semester.'''
    #This part of the function evaluates the student’s participation.
    if int(participation) >= 8:
        par = ("fantastic")
        par_sent = (f" Every day {pro1} comes to class ready to learn, and challenges the students around {pro2}")
    if 8 > int(participation) >= 5:
        par = ("good")
        par_sent = (f" Every day {pro1} comes to class and does his work efficiently")
    if 5 > int(participation) >= 3:
        par = ("alright")
        par_sent = (f" I think that it would be very benifitial to {pro2} if {pro1} puts a little bit more effort into learning how to code")
    if int(participation) < 3:
        par = ("bad")
        par_sent = (f" I feel that {pro1} comes to class not ready to learn and distracts the students around {pro2}") 
    #This part of the function evaluates the student’s motivation.
    if int(motivation) >= 8:
        mot = ("extremely")
        mot_sent = (f" Even after {pro1} finishes the assigned work, {first} likes to do extra work")
    if 8 > int(motivation) >= 5:
        mot = ("very")
        mot_sent = (f" {first} gets all of the work done and occasionally does extra work")
    if 5 > int(motivation) >= 3:
        mot = ("some what")
        mot_sent = (f" {first} usually gets all of the work done on time, and if {pro1} needs and extention {pro1} always askes")
    if int(motivation) < 3:
        mot = ("not")
        mot_sent = (f" {first} rarely gets work done, and forgets to submit homework and problem sets")
    #This part of the function evaluates the student’s attendance.
    if int(attendance) >= 8:
        att = ("1 or 2")
        attend = (f", and {pro1} always tells me when {pro1} won't be there.")
    if 8 > int(attendance) >= 5:
        att = ("4 or 5")
        attend = (f" But {pro1} always tells me when {pro2} won't be there.")
    if 5 > int(attendance) >= 3:
        att = ("a lot of")
        attend = (f", and {pro1} needs to communicate with me more if {pro2} will be missing class.")
    if int(attendance) < 3:
        att = ("too many days")
        attend = (f" {pro1} is never there and is very far behind.")
    #This returnes the full paragraph.
    return(f"{first} has {par} participation and is {mot} motivated.{par_sent}.{mot_sent}. {pro1.title()} has missed {att} classes{attend}\n")



#Created this function to be used by grade_calc function to find correct infedinite article for grade.
def an_vs_a_func(letter_grade):
    '''This function takes input of a letter grade and returns the correct an/a indefinite article'''
    if 'A' in letter_grade:
            return "an"
    if 'B' in letter_grade:
            return "a"
    if 'C' in letter_grade:
            return "a"
    if 'D' in letter_grade:
            return "a"
    if 'F' in letter_grade:
            return 'an'

#One of our many functions that summarizes student performance based on test and problem set grades.
def grade_summary(first, test, problem_sets, pronoun1):
    '''Takes in firstname, test score, problem set score, and he/her/they pronoun. Returns string of grade summary based on test and problem set scores.'''
    score_sentence=[]
    score_sentence.append(first.capitalize())
    if int(test) > 90:
        score_sentence.append(" has been performing well on our tests. ")
    if 80 < int(test) < 91:
        score_sentence.append(" has performed pretty average on tests. ")
    if int(test) < 81:
        score_sentence.append(" needs to study more for tests. ")
    score_sentence.append(pronoun1.capitalize())
    if int(problem_sets) > 90:
        score_sentence.append(" has completed all problem sets this year. ")
    if 80 < int(problem_sets) < 91:
        score_sentence.append(" has completed most of the problem sets this year. ")
    if int(problem_sets) < 81:
        score_sentence.append(" has not been doing well on problem sets. ")
    return ''.join(score_sentence)
#Tells student if they need to improve or not based on their grade. Also summarizes their final letter grade
def grade_calc(letter_grade, first, final_grade, pronoun1, pronoun2):
    '''Takes in student letter grade, name, final grade, he/she pronoun, and his/her pronoun.'''
    grade_sentence=[]
    grade_sentence.append(first.capitalize())
    if int(final_grade) > 90:
        grade_sentence.append(" has been doing really well this year. ")
    if 80 < int(final_grade) < 91:
        grade_sentence.append(" has had an average performance this year. ")
    if int(final_grade) < 81:
        grade_sentence.append(" needs to start performing better. ")
    grade_sentence.append(f"{pronoun1.capitalize()} currently has {an_vs_a_func(letter_grade)} {letter_grade} in the class.")
    return ''.join(grade_sentence)
#Code to loop through rows in comments and create text files based on student name. If file already exists it is updated. Strings from previous data are written to file one at a time with appropriate formatting.
for info in comments:
    filename = (info[1] + ".txt")
    name_file = open(filename, "w")
    name_file.write(info[1] + "'s " + info[4] + " block comments:" + "\n" + "\n" )
    name_file.write(course + '\n' + '\n')
    work = work_ethic(info[1], info[2], info[3], info[5], info[-2], info[-1])
    name_file.write(work + '\n')
    grade = grade_summary(info[1], info[6], info[7], info[2])
    name_file.write(grade)
    calc = grade_calc(info[9], info[1], info[8], info[2], info[3])
    name_file.write(calc)
    name_file.close()
```
